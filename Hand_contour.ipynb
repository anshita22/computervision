```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Hands..jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)    # Converting Image to Grayscale
blurred = cv2.GaussianBlur(gray, (15,15), 0)  #  Blurring the Image
edged = cv2.Canny(blurred, 30, 180)           #   Detecting Edgesgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)    # Converting Image to Grayscale

plt.imshow(edged)
plt.show()
```

```python
_,cnts, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
hands = img.copy()
cv2.drawContours(hands, cnts, -1, (0, 0, 255), 5)
plt.imshow(hands)
plt.show()
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
