```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img= cv2.imread('d.png')
```

```python
plt.imshow(img)
plt.show()
```

```python
#blur the image
blurimage= cv2.GaussianBlur(img, (7, 7), 0)
plt.imshow(blurimage)
plt.show()
```

```python
#getting edges 
edge = cv2.Canny(blurimage, 120, 250)
plt.imshow(edge)
plt.show()
```
